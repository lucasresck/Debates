```python
import pandas as pd
import numpy as np
import random
import math
```

```python
nderodadas = 3 #Fixo
ndeduplas = 32 #Pode variar
ndesalas = ndeduplas / 4
```

```python
def leitura():
    
    #Leitura dos juízes e das duplas

    duplas = pd.read_excel("GV Debate.xlsx", sheet_name = "Duplas")
    juizes = pd.read_excel("GV Debate.xlsx", sheet_name = "Juízes")
    
    #Leitura das salas
    
    salas = [0]
    for sala in range(1, int(ndesalas) * nderodadas + 1):
        salas.append(pd.read_excel("GV Debate.xlsx", sheet_name = str(sala)))
    
    #Leitura das ocorrencias        
        
    ndewings = juizes["#"][juizes["#"] < 200].size
    ndechairs = juizes["#"][juizes["#"] >= 200].size

    wingwing = np.zeros(ndewings ** 2).reshape(ndewings, ndewings)
    wingchair = np.zeros(ndewings * ndechairs).reshape(ndewings, ndechairs)
    wingdupla = np.zeros(ndewings * ndeduplas).reshape(ndewings, ndeduplas)
    
    novasduplas = np.array([])
    novoschairs = np.array([])

    for sala in range(1, int(ndesalas) * nderodadas + 1):
        duplasnestasala = salas[sala]["#"]
        duplasnestasala = duplasnestasala.dropna()
        juizesnestasala = salas[sala]["Juízes"]
        juizesnestasala = juizesnestasala.dropna()
        chairsnestasala = juizesnestasala[juizesnestasala >= 200].values
        wingsnestasala = juizesnestasala[juizesnestasala < 200].values
        
        if len(wingsnestasala) == 0: #Leitura das duplas da nova rodada (já adicionadas na planilha)
            novasduplas = np.append(novasduplas, duplasnestasala.values)
            novoschairs = np.append(novoschairs, chairsnestasala)

        if len(wingsnestasala) > 0:
            for wing1 in range(0, len(wingsnestasala)):
                for wing2 in range(wing1 + 1, len(wingsnestasala)):
                    if wingsnestasala[wing1] > wingsnestasala[wing2]:
                        wingwing[int(wingsnestasala[wing1]) - 100 - 1][int(wingsnestasala[wing2]) - 100 - 1] = wingwing[int(wingsnestasala[wing1]) - 100 - 1][int(wingsnestasala[wing2]) - 100 - 1] + 1
                    else:
                        wingwing[int(wingsnestasala[wing2]) - 100 - 1][int(wingsnestasala[wing1]) - 100 - 1] = wingwing[int(wingsnestasala[wing2]) - 100 - 1][int(wingsnestasala[wing1]) - 100 - 1] + 1


        if len(wingsnestasala) > 0 and len(chairsnestasala) > 0:
            for wing in range(0, len(wingsnestasala)):
                wingchair[int(wingsnestasala[wing]) - 100 - 1][int(chairsnestasala[0]) - 200 - 1] = wingchair[int(wingsnestasala[wing]) - 100 - 1][int(chairsnestasala[0]) - 200 - 1] + 1

        if len(wingsnestasala) > 0 and len(duplasnestasala) > 0:
            for wing in range(0, len(wingsnestasala)):
                for dupla in range(0, len(duplasnestasala)):
                    wingdupla[int(wingsnestasala[wing]) - 100 - 1][int(duplasnestasala[dupla]) - 1] = wingdupla[int(wingsnestasala[wing]) - 100 - 1][int(duplasnestasala[dupla]) - 1] + 1
            
    #Leitura das restrições        
            
    restricoes = pd.read_excel("GV Debate.xlsx", sheet_name = "Restrição de juízes")
        
    return duplas, juizes, salas, wingwing, wingchair, wingdupla, restricoes, novasduplas, novoschairs, ndewings
```

```python
#Sorteio dos juizes

def sorteioJuizes(juizes):

    sortwings = np.array(juizes["#"][juizes["#"] < 200].values)
    random.shuffle(sortwings)
    
    return sortwings
```

```python
def testes(duplas, juizes, salas, wingwing, wingchair, wingdupla, restricoes, novasduplas, novoschairs, ndewings):
    configuracoes = []
    restrito = False
    for vez in range(0, 10000):
        sortwings = sorteioJuizes(juizes)
        qtdewingssala = math.floor(ndewings / ndesalas)
        sobrawings = int(ndewings - qtdewingssala * ndesalas)

        wingwingaux = np.array(wingwing)
        wingchairaux = np.array(wingchair)
        wingduplaaux = np.array(wingdupla)
        winganterior = 0

        for sala in range(1, int(ndesalas) + 1):
            if sala == int(ndesalas):
                wingsnestasala = sortwings[winganterior:]
            elif sala <= sobrawings:
                wingsnestasala = sortwings[winganterior:(winganterior + qtdewingssala + 1)]
                winganterior = winganterior + qtdewingssala + 1
            else:
                wingsnestasala = sortwings[winganterior:(winganterior + qtdewingssala)]
                winganterior = winganterior + qtdewingssala
            duplasnestasala = novasduplas[(0 + 4 * (sala - 1)):(4 + 4 * (sala - 1))]
            chairsnestasala = novoschairs[sala - 1]

            for restricao in range(0, len(restricoes)):
                if restricoes['# dupla'][restricao] in duplasnestasala and restricoes['# juíz'][restricao] in wingsnestasala:
                    restrito = True

            #wingwing

            for wing1 in range(0, len(wingsnestasala)):
                for wing2 in range(wing1 + 1, len(wingsnestasala)):
                    if wingsnestasala[wing1] > wingsnestasala[wing2]:
                        wingwingaux[int(wingsnestasala[wing1]) - 100 - 1][int(wingsnestasala[wing2]) - 100 - 1] = wingwingaux[int(wingsnestasala[wing1]) - 100 - 1][int(wingsnestasala[wing2]) - 100 - 1] + 1
                    else:
                        wingwingaux[int(wingsnestasala[wing2]) - 100 - 1][int(wingsnestasala[wing1]) - 100 - 1] = wingwingaux[int(wingsnestasala[wing2]) - 100 - 1][int(wingsnestasala[wing1]) - 100 - 1] + 1

            #wingchair

            for wing in range(0, len(wingsnestasala)):
                wingchairaux[int(wingsnestasala[wing]) - 100 - 1][int(chairsnestasala) - 200 - 1] = wingchairaux[int(wingsnestasala[wing]) - 100 - 1][int(chairsnestasala) - 200 - 1] + 1

            #wingdupla

            for wing in range(0, len(wingsnestasala)):
                for dupla in range(0, len(duplasnestasala)):
                    wingduplaaux[int(wingsnestasala[wing]) - 100 - 1][int(duplasnestasala[dupla]) - 1] = wingduplaaux[int(wingsnestasala[wing]) - 100 - 1][int(duplasnestasala[dupla]) - 1] + 1

        wingwingaux[wingwingaux != 0] = wingwingaux[wingwingaux != 0] - 1
        wingchairaux[wingchairaux != 0] = wingchairaux[wingchairaux != 0] - 1
        wingduplaaux[wingduplaaux != 0] = wingduplaaux[wingduplaaux != 0] - 1

        somarepquad = sum(sum(wingwingaux ** 2)) + sum(sum(wingchairaux ** 2)) + sum(sum(wingduplaaux ** 2))
        if not restrito:
            configuracoes.append([sortwings, somarepquad])
        else:
            restrito = False
    
    return configuracoes
```

```python
def orgwings(configuracoes):
    melhorconfig = min(configuracoes, key = lambda configuracao: configuracao[1])
    print("Soma das repetições ao quadrado: {}".format(melhorconfig[1]))
    
    ndewings = len(melhorconfig[0])
    ordwings = melhorconfig[0]    
    
    qtdewingssala = math.floor(ndewings / ndesalas)
    sobrawings = int(ndewings - qtdewingssala * ndesalas)
    winganterior = 0
    wingsfinal = {}
    
    for sala in range(1, int(ndesalas) + 1):
        if sala == int(ndesalas):
            wingsfinal.update({"Sala {}".format(sala): list(ordwings[winganterior:])})
        elif sala <= sobrawings:
            wingsfinal.update({"Sala {}".format(sala): list(ordwings[winganterior:(winganterior + qtdewingssala + 1)])})
            winganterior = winganterior + qtdewingssala + 1
        else:
            wingsfinal.update({"Sala {}".format(sala): list(ordwings[winganterior:(winganterior + qtdewingssala)])})
            winganterior = winganterior + qtdewingssala
    
    print(wingsfinal)
    return wingsfinal
```

```python
duplas, juizes, salas, wingwing, wingchair, wingdupla, restricoes, novasduplas, novoschairs, ndewings = leitura()
```

```python
configuracoes = testes(duplas, juizes, salas, wingwing, wingchair, wingdupla, restricoes, novasduplas, novoschairs, ndewings)
```

```python
wingsfinal = orgwings(configuracoes)
```
